```python
import pandas as pd

# Load the Titanic dataset
Titanic = "/content/titanic.csv"
titanic_df = pd.read_csv(Titanic)

# Display the first few rows
print("Original Titanic Dataset:")
print(titanic_df.head())

# Fill missing values
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)  # Fill missing ages with median
titanic_df['Fare'].fillna(titanic_df['Fare'].mean(), inplace=True)  # Fill missing fares with mean

# Add a Survival Cost column (Fare / Survival status)
titanic_df['Survival_Cost'] = titanic_df['Fare'] / (titanic_df['Survived'] + 1)  # Avoid division by 0

# Display the cleaned dataset
print("\nCleaned Titanic Dataset:")
print(titanic_df.head())

# Export the cleaned dataset
output_path = "Cleaned_Titanic.csv"
titanic_df.to_csv(output_path, index=False)

print(f"\nCleaned Titanic dataset saved to: {output_path}")

# Bonus Practical

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

Titanic = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_df = pd.read_csv(Titanic)

# Survival rate by class and gender
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=titanic_df)
plt.title("Survival Rate by Class and Gender")
plt.show()

# Age distribution
sns.histplot(titanic_df['Age'], kde=True, bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
